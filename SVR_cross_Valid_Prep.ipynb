```
import nltk
import pandas as pd
import numpy as np
from sklearn.model_selection import KFold 
from sklearn.svm import SVR, NuSVR
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
import joblib
```

```
training_data = pd.read_csv('./csv_files/training_no_emoji.csv')
```

```
kf = KFold(n_splits=4, random_state=42, shuffle=True)
```

```
t = 0
for train_index, test_index in kf.split(training_data):
    X_train, X_test = training_data['Text'][train_index], training_data['Text'][test_index]
    y_train , y_test = training_data['Lat'][train_index] , training_data['Lat'][test_index]
    with open('./code/texts/train_text' + str(t) + '.txt', "w") as file:
      for text in X_train:
        file.write(text + "\n")
    with open('./code/texts/train_lat' + str(t) + '.txt', "w") as file:
      for lat in y_train:
        file.write(str(lat) + "\n")
    with open('./code/texts/test_text' + str(t) + '.txt', "w") as file:
      for text in X_test:
        file.write(text + "\n")
    with open('./code/texts/test_lat' + str(t) + '.txt', "w") as file:
      for lat in y_test:
        file.write(str(lat) + "\n")
    t += 1
```

```
with open('./texts/train_lat0.txt', "r") as file:
  mat = [[float(num) for num in line.split(' ') if num != '\n'] for line in file]
training_lats0 = np.array(mat)
```

```
dict_data = np.load('./kernels/train_kernel0.npz')
training_kernel0 = dict_data['arr_0']
```

```
nuSvr = NuSVR(kernel='precomputed', C = 10, cache_size = 2048, max_iter = 500000, verbose=True)
```

```
nuSvr.fit(training_kernel0, training_lats0.ravel())
```

```
joblib.dump(nuSvr, 'crossValidNuSvr.pkl')
```
